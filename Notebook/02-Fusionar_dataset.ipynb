# Fusionar conjunto de datos

```python
dataset = concat_csv()
dataset.head(3) # muestra de las primeras 3 lineas del dataset fusionado
```

```python
dataset.columns = ['fecha', 'forecast', 'lower_forecast', 'upper_forecast', 'trend', 'prov_re', 'prov_en', 'envios', 'festivo']
dataset[['forecast', 'lower_forecast', 'upper_forecast', 'trend', 'envios', 'festivo']] = dataset[['forecast', 'lower_forecast', 'upper_forecast', 'trend', 'envios', 'festivo']].astype("Int64")
dataset.head(3)
```

```python
df_regiones = pd.read_csv('../Data/prov_reg.csv')
df_regiones.sample(3)
```

```python
dataset = pd.merge(dataset, df_regiones, how='left', left_on='prov_re', right_on='prov').rename(columns = {'reg': 'reg_re', 'coordenadas_prov': 'coor_re'})
dataset = pd.merge(dataset, df_regiones, how='left', left_on='prov_en', right_on='prov').rename(columns = {'reg': 'reg_en', 'coordenadas_prov': 'coor_en'})
dataset.drop(['prov_x', 'prov_y'], axis=1, inplace=True)
```

```python
dataset = dataset[['fecha', 'prov_re', 'coor_re','reg_re','prov_en', 'coor_en','reg_en', 'festivo', 'envios', 'forecast', 'lower_forecast', 'upper_forecast', 'trend']]
dataset.sample(3)
```
